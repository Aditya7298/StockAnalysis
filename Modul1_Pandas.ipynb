### Query 1.4
In a separate array , calculate the monthwise VWAP (Volume Weighted Average Price ) of the stock. 
( VWAP = sum(price*volume)/sum(volume) ) 
To know more about VWAP , visit - VWAP definition

### Query 1.5
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def average_price_last_n_days(n,df):
    start = len(df) - n
    end = len(df) + 1
    price = np.sum(df[start:end][['Close Price']])
    return float(price/n)

def profit_loss_change(n,df):
    last_day = len(df) - n
    curr_price = df.iloc[len(df)-1][['Close Price']]
    last_price = df.iloc[last_day][['Close Price']]
    change = curr_price - last_price
    per_change = change/curr_price
    return float(per_change*100)

print("Average Price and % change fo last 1 week are {a:.2f} and {b:.2f}% respectively".format(a = average_price_last_n_days(7,df),b = profit_loss_change(7,df)) )
print("Average Price and % change fo last 2 weeks are {a:.2f} and {b:.2f}% respectively".format(a = average_price_last_n_days(14,df),b = profit_loss_change(14,df)) )
print("Average Price and % change fo last 1 month are {a:.2f} and {b:.2f}% respectively".format(a = average_price_last_n_days(31,df),b = profit_loss_change(31,df)) )
print("Average Price and % change fo last 1 year are {a:.2f} and {b:.2f}% respectively".format(a = average_price_last_n_days(365,df),b = profit_loss_change(365,df)) )
```

Add a column 'Day_Perc_Change' where the values are the daily change in percentages i.e. the percentage change between 2 consecutive day's closing prices. ### Query 1.7
 Add another column 'Trend' whose values are:
'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
'Positive' for 'Day_Perc_Change' in between 1 and 3
'Negative' for 'Day_Perc_Change' in between -1 and -3
'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
'Among top losers' for 'Day_Perc_Change' in between -3 and -7
'Bull run' for 'Day_Perc_Change' >7
'Bear drop' for 'Day_Perc_Change' <-7

```python
df['Trend'] = df['Day_Perc_Change'].astype(str)

def return_trend(n):
    
    n = float(n)
    
    if(-.5 <= n and n < .5):
        return "Slight or No Change"
    
    elif(.5 <= n and n < 1):
        return "Slight Positive"
    
    elif(-1 <= n and n < -.5):
        return "Slight Negative"
    
    elif(1 <= n and n < 3):
        return "Positive"
    
    elif(-3 <= n and n < -1):
        return "Negative"
    
    elif(3 <= n and n < 7):
        return "Among Top gainers"
    
    elif(-7 <= n and n < -3):
        return "Among Top losers"
    
    elif(n >= 7):
        return "Bull Run"
    
    elif(n <= -7):
        return "Bear Drop"
    
    else:
        return ""
    

df['Trend'] = df['Trend'].apply(return_trend)
df['Trend']
```
